```python
i = 45
dim = (512,512)

img_height = cv2.imread('./dataset_01_lowRes/test_05_height (%i).png'%i)[:,:,0]
img_rgb = cv2.imread('./dataset_01_lowRes/test_05_rgb (%i).png'%i)
img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)

img_height_resize = cv2.resize(img_height, dim, interpolation = cv2.INTER_AREA)
img_rgb_resize = cv2.resize(img_rgb, dim, interpolation = cv2.INTER_AREA)
img_stack = np.dstack([img_rgb_resize,img_height_resize])

print(img_resize.shape)
plt.imshow(img_stack)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20,20))

axs[0].imshow(img_rgb_resize)
axs[1].imshow(img_height_resize, cmap='gray')
axs[2].imshow(img_stack)
```

```python
i = 45
dim = (512,512)

for i in range(1,50):
    try:
        
        img_height = cv2.imread('./dataset_01_lowRes/test_05_height (%i).png'%i)[:,:,0]
        img_rgb = cv2.imread('./dataset_01_lowRes/test_05_rgb (%i).png'%i)

        img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)

        img_height_resize = cv2.resize(img_height, dim, interpolation = cv2.INTER_AREA)
        img_rgb_resize = cv2.resize(img_rgb, dim, interpolation = cv2.INTER_AREA)
        img_stack = np.dstack([img_rgb_resize,img_height_resize])

        cv2.imwrite("merged/%i.png"%i, img_stack)
        
    except TypeError:
        pass
```
